```python
import numpy as np
import pandas as pd
from matplotlib import pyplot 
```

```python
class linearRegression:
    def data(self,filepath):
        df=pd.read_csv(filepath)
        df.dropna(inplace=True)
        return df
        
```

```python
from matplotlib import pyplot as plt

def costfn(column1, column2, w, b):
    range1 = len(column1)
    # compute total cost (mean squared error)
    total_cost = sum((w*column1[i] - column2[i])**2 for i in range(range1)) / range1
    print(f"Total cost for w={w}, b={b} is {total_cost}")
    return total_cost

values = [1,2,3,4,5,6,7]
output = [2,4,6,8,10,12,14]  # fixed the 110 typo

best_cost = float('inf')
best_params = (0,0)

for i in range(10):   # smaller range for testing
    for j in range(10):
        c = costfn(values, output, i, j)
        if c < best_cost:
            best_cost = c
            best_params = (i, j)

print(f"Lowest cost is {best_cost} at w={best_params[0]}, b={best_params[1]}")
a=int(input("Enter the value you want to get value from our linear model."))
print(f"The required result is :{best_params[0]*a+best_params[1]}")
    
```
